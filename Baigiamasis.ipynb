```python
df.groupby('Product line').agg({'Quantity': np.sum,
                                'Total' : np.sum,
                                'Tax 5%' : np.sum,
                                'cogs'  : np.sum})
```

```python
sns.countplot(y = 'Product line', data = df, hue = 'Gender')
plt.xlabel('Genders')
plt.title('Customers', fontsize=16)
plt.style.use('dark_background')
plt.legend(loc = 'upper left')
```

```python
df.groupby('Gender').agg({'Quantity': np.sum,
                          'Total'  : np.sum})
```

```python
sns.countplot(y = 'Product line', data = df, hue = 'City')
plt.xlabel('Invoices')
plt.title('Sales', fontsize=16)
plt.legend(loc = 'upper left')
```

```python
sns.displot(df['Payment'], color = 'orange')
plt.style.use('dark_background')
plt.xlabel('Payment type')
plt.ylabel('Payments')
```
